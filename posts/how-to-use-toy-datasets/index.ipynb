# sklearn.datasets

```python
import pandas as pd
import numpy as np
import sklearn.datasets
```

## boston

```python
dataset = sklearn.datasets.load_boston()
df = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])
y = dataset['target']
print(df.shape, y.shape)
display(df.head())
display(y[:10])
```

## diabets

```python
dataset = sklearn.datasets.load_diabetes()
df = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])
y = dataset['target']
print(df.shape, y.shape)
display(df.head())
display(y[:10])
```

## breast_cancer

```python
dataset = sklearn.datasets.load_breast_cancer()
df = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])
y = dataset['target']
print(df.shape, y.shape)
display(df.head())
display(np.unique(y))
```

## iris

```python
dataset = sklearn.datasets.load_iris()
df = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])
y = dataset['target']
print(df.shape, y.shape)
display(df.head())
display(np.unique(y))
```
